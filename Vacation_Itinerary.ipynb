```python
# Import the dependencies.
import pandas as pd
import gmaps
import requests
# Import the API key.
from config import g_key
```

```python
# Configure gmaps to use your Google API key.
gmaps.configure(api_key=g_key)
```

```python
# Store the CSV you saved created in part one into a DataFrame.
city_data_df = pd.read_csv("weather_data/WeatherPy_vacation.csv")
city_data_df
```

```python
#initiatie list of cities to travel to
cities_traveled = ['Sabaudia','Bose','Saint-Francois','Saint-Denis']
```

```python
#filter df down to those cities
cities_traveled_df=city_data_df[city_data_df["City"].isin(cities_traveled)]
cities_traveled_df
```

```python
# create individual dfs for each city
Sabaudia_df=cities_traveled_df[cities_traveled_df["City"]=='Sabaudia']
Bose_df=cities_traveled_df[cities_traveled_df["City"]=='Bose']

Saint_Francois_df= cities_traveled_df[cities_traveled_df["City"]=='Saint-Francois']
Saint_Denis_df =cities_traveled_df[cities_traveled_df["City"]=='Saint-Denis']
```

```python
#extract lats to lists
Sabaudia_lat = Sabaudia_df['Lat'].tolist()
Bose_lat = Bose_df["Lat"].tolist()
Saint_Francois_lat = Saint_Francois_df["Lat"].tolist()
Saint_Denis_lat = Saint_Denis_df["Lat"].tolist()

# extract lngs to lists
Sabaudia_lng = Sabaudia_df['Lng'].tolist()
Bose_lng = Bose_df["Lng"].tolist()
Saint_Francois_lng = Saint_Francois_df["Lng"].tolist()
Saint_Denis_lng =Saint_Denis_df["Lng"].tolist()

#combine lists
Sabaudia_lat_lng = Sabaudia_lat + Sabaudia_lng
Bose_lat_lng = Bose_lat + Bose_lng
Saint_Francois_lat_lng =Saint_Francois_lat + Saint_Francois_lng
Saint_Denis_lat_lng =Saint_Denis_lat + Saint_Denis_lng

Sabaudia_lat_lng = tuple(Sabaudia_lat_lng)
Bose_lat_lng = tuple(Bose_lat_lng)
Saint_Francois_lat_lng = tuple(Saint_Francois_lat_lng)
Saint_Denis_lat_lng = tuple(Saint_Denis_lat_lng)
```

```python
#generate map with directions layer
fig=gmaps.figure()
road_trip=gmaps.directions_layer(Sabaudia_lat_lng,Saint_Denis_lat_lng,waypoints=[Bose_lat_lng,Saint_Francois_lat_lng], travel_mode='DRIVING')
fig.add_layer(road_trip)

fig
```

```python
# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in cities_traveled_df.iterrows()]
```
